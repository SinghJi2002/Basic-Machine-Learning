### Simple Linear Regression

The goal of a SML model is as simple as it gets. It has a input database, it just need to draw a line such that the sum of distance of each datapoint used in creating that line from the line (error) is the least. 

The general equation that we need to derive in Simple Linear Equation is 

### <center>F(x)=wx+b</center>

Here __w__ is the slope and __b__ is the intercept. These are the actual thing we need to derive.

Now how to be find the correct w and b? . That is, how do we derive the errors to compute for which value of of w and b, the error encountered is the least. This is done via the help of __Cost Function__. For this case we are going to concentrate of only one of the famous cost functions, that is __Mean Squared Error__ or __MSE__. Though all this sounds good and over, one issue here is the methodology. Its not feasable for us to be put in a situation where we are manually computing the best possible cost/loss via hit and trail method and hence we require a convergence algorithm that can do this for us. This convergence algorithm is known as __Gradient Descent__.

The general steps of applying gradient decent involves:
* __Step1__: Take any given function. Assume initial values of parameters.
* __Step2__: Keep changing the values of the parameter via algorithm to lower the value of J,
* __Step3__: Keep repeating until minimum value of J/convergence is attained.
