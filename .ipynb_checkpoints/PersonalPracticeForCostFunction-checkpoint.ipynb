# Machine Learning Optional Lab 2

Let optimize our learning a bit. Let use the following model for value derievation.

                                         f_w=w*x
Now here we are going to apply the concept of cost function. To know more about it please read the notes. 

The sample dataset that we are going to examine are as follow:

House Size(in 1000 sq.feets) : House Price(in 100000 dollars)
            
            2                           5
            
            5                           8

In example ahead we are going to see two different graphs. A normal plot between y hat and x like the previous case. We will also see a plot between J and w.

```python
#Creating functions for 
x_values_HouseSize=np.array([2,5])
y_values_HousePrize=np.array([5,8])
```

```python
x_values_HouseSize.shape[0]
```

```python
#So the focus of this module is on understanding the working of J v/s w graph in case of single and multivariable setting.
#Lets declare a 10 element list that has a few values w we want to assign to w and find J in each case.

#Lets declare a function to find the values of J
def W_value():
    values=[]
    for j in range(0,11,1):
        Wvalue=int(input("Enter Value Of W of your choice"))
        values.append(Wvalue)
    return(np.array(values))


def J_value(OrigValue,Xvalues,Wvalues):
    valuesOfJ=[]
    count=0
    while(count<11):
        prediction=[]
        intSum=0
        for i in range(0,Xvalues.shape[0],1):
            PredictValueI=Wvalues[count]*Xvalues[i]
            prediction.append(PredictValueI)
        for k in range(0,Xvalues.shape[0],1):
            Jvalue=((prediction[k])**2)-((OrigValue[k])**2)
            intSum=intSum+Jvalue
        realValueOfJ=intSum/4
        valuesOfJ.append(realValueOfJ)
        count=count+1
    return(np.array(valuesOfJ))
            
valuesOfW=W_value()
valuesOfJ=J_value(x_values_HouseSize,y_values_HousePrize,valuesOfW)
```

```python
valuesOfJ
```

```python
valuesOfW
```

```python
#Lets plot this data and see what comes out of it.
plt.scatter(valuesOfW,valuesOfJ,marker='x',color='r')
plt.plot(valuesOfW,valuesOfJ,color='r')
plt.xlabel('w')
plt.ylabel('J')
plt.title("J v/s w")
```

We clearly see the plot above between J and w. From above plot we need to select the value of w at which the least value of J above J=0 is obtained. Hence w=1, hence that will fit the graph perfectly. Lets see it in application.

```python
y_predicts=[]
w=1
for i in range(0,2,1):
    newY=w*y_values_HousePrize[i]
    y_predicts.append(newY)
y_hat=np.array(y_predicts)
```

```python
plt.scatter(x_values_HouseSize,y_values_HousePrize,marker='x',color='b',label='Original Values')
plt.plot(x_values_HouseSize,y_hat,color='r',label='Predicted Values')
plt.legend()
plt.xlabel("House Size")
plt.ylabel("House Price")
```

# 3D Plotting
Let change our focus, to more of a fullform of linear regression equation that is
    __f=wx+b__

The default x and y datasets are as follows,
    
    House Size(1000s sq feet)                 House Price(100000s Dollars)
            
            2                                      5
            
            5                                      8

Here we have 2 changable parameters w and b, whose optimum value we need to determine to find the best possible fit.

```python
#Let make array of concerned datasets
x_axis=np.array([1.0, 2.0])
y_axis=np.array([300.0,500.0])
```

```python
#Lets make a 3 functions. One functions collects the values of w, the other stores the values of b, while the other stores the 
#values of y hat.

def W_value():
    values1=[]
    for j in range(0,11,1):
        Wvalue=int(input("Enter Value Of W of your choice"))
        values1.append(Wvalue)
    return(np.array(values1))


def B_value():
    values2=[]
    for j in range(0,11,1):
        Wvalue=int(input("Enter Value Of b of your choice"))
        values2.append(Wvalue)
    return(np.array(values2))


def J_value(OrigValue,Xvalues,Wvalues,Bvalues):
    valuesOfJ=[]
    count=0
    while(count<11):
        prediction=[]
        intSum=0
        for i in range(0,Xvalues.shape[0],1):
            PredictValueI=Wvalues[count]*Xvalues[i]+Bvalues[count]
            prediction.append(PredictValueI)
        for k in range(0,Xvalues.shape[0],1):
            Jvalue=((prediction[k])**2)-((OrigValue[k])**2)
            intSum=intSum+Jvalue
        realValueOfJ=intSum/4
        valuesOfJ.append(realValueOfJ)
        count=count+1
    return(np.array(valuesOfJ))
            
valuesOfW=W_value()
valuesOfB=B_value()
W,B=np.meshgrid(valuesOfW,valuesOfB)
valuesOfJ=J_value(x_values_HouseSize,y_values_HousePrize,W,B)
```
