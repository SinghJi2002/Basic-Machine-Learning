```python
import numpy as np
import matplotlib as plt
import copy
```

```python
#Lets first create a training set which we are going to work with
x_train=np.array([[0.5, 1.5], [1,1], [1.5, 0.5], [3, 0.5], [2, 2], [1, 2.5]])
y_train=np.array([0, 0, 0, 1, 1, 1])
```

```python
#let first declare the parameters that we are going to use hare.
b=0
w=np.zeros(x_train.shape[1])
```

```python
#Lets now create the our function which will formulate the working model(Sigmoid Function)
def model(X,Y,w,b):
    model=np.dot(w,X)+b
    sigmoid=1/(1+np.exp(-model))
    return(sigmoid)
```

```python
#Lets now create our function for calculating the cost
def cost(X,Y,w,b):
    intermediate_Cost=0
    for i in range(0,X.shape[0],1):
        prediction=model(X[i],Y,w,b)
        intermediate_Cost=intermediate_Cost+((-Y[i])*np.log(prediction))+((1-Y[i])*np.log(1-prediction))
    cost=intermediate_Cost/X.shape[0]
    return(cost)
```

```python
#Let start with implementing the gradient decent. But first we implement Gradients
def gradients(X,Y,w,b):
    dj_db=0
    dj_dw=np.zeros(X.shape[1])
    intermediate_sum=0
    for i in range(0,X.shape[0],1):
        prediction=model(X[i],Y,w,b)
        intermediate=prediction-Y[i]
        for j in range(0,X.shape[1],1):
            intermediate_sum=intermediate_sum+intermediate*X[i,j]
        dj_db=dj_db+prediction
    dj_dw=intermediate_sum/X.shape[0]
    dj_db=dj_db/X.shape[0]
    return(dj_dw,dj_db)
```

```python
#Lets now finally calculate the decent
def decent(X,Y,w,b,alpha,iterations):
    J_history=[]
    J_history.append(cost(X,Y,w,b))
    for i in range(0,iterations,1):
        dj_dw,dj_db=gradients(X,Y,w,b)
        w=w-(alpha*dj_dw)
        b=b-(alpha*dj_db)
        J_history.append(cost(X,Y,w,b))
    return(J_history,w,b)        
```

```python
a,b,c=decent(x_train,y_train,w,b,0.01,1000)
```
